### ğŸ“ ìˆ™ì œ 2: ë‚˜ë§Œì˜ "ë§¤ìˆ˜/ë§¤ë„ ê·œì¹™" ë§Œë“¤ê¸° (Logic)

```python
### ìˆ™ì œ: ê³¨ë“ í¬ë¡œìŠ¤(Golden Cross) ì „ëµ ë§Œë“¤ê¸°
# ê·œì¹™: 5ì¼ì„ ì´ 20ì¼ì„ ë³´ë‹¤ ìœ„ë¡œ ì˜¬ë¼ê°€ë©´ 'ë§¤ìˆ˜(1)', ì•„ë‹ˆë©´ 'ëŒ€ê¸°(0)'

df['Signal'] = 0

# ì•„ë˜ ë¹ˆì¹¸ì— ì¡°ê±´ì‹ì„ ì™„ì„±í•´ë³´ì„¸ìš”. (np.whereë¥¼ ì“°ë©´ ì‰½ìŠµë‹ˆë‹¤)
import numpy as np
df['Signal'] = np.where(df['MA5'] > df['MA20'], 1, 0)

# [í™•ì¸] ìµœê·¼ 10ì¼ê°„ì˜ ì‹ í˜¸ ì¶œë ¥
print(df['Signal'].tail(10))

```

### ğŸ“ ìˆ™ì œ 3: AIì—ê²Œ "ë‚´ì¼ ê°€ê²©" ë¬¼ì–´ë³´ê¸° (Predict)

**ëª©í‘œ:** ê³¼ê±°ì˜ ë°ì´í„°(MA5, MA20)ë¥¼ ë³´ê³  AIê°€ ë‚´ì¼ì˜ ê°€ê²©ì„ ë§ì¶”ë„ë¡ í•™ìŠµì‹œí‚µë‹ˆë‹¤.

```python
from sklearn.ensemble import RandomForestRegressor # ì´ë²ˆì—” ë” ì‰¬ìš´ ëœë¤í¬ë ˆìŠ¤íŠ¸!

# [ì¤€ë¹„] ë‚´ì¼ ê°€ê²©(Target) ë§Œë“¤ê¸°
df['Target'] = df['Close'].shift(-1)
df = df.dropna()

### ìˆ™ì œ: AI í•™ìŠµì‹œí‚¤ê¸°
# 1. í•™ìŠµì— ì“¸ ì¬ë£Œ(X)ë¥¼ ì„ íƒí•˜ì„¸ìš”. (ì˜ˆ: MA5, MA20)
# 2. ë§ì¶œ ì •ë‹µ(y)ì„ ì„ íƒí•˜ì„¸ìš”. (ì˜ˆ: Target)

X = df[['MA5', 'MA20']]
y = df['Target']

# ëª¨ë¸ ìƒì„± ë° í•™ìŠµ
model = RandomForestRegressor(n_estimators=50)
model.fit(X, y)

# ë‚´ì¼ì˜ ì˜ˆìƒ ê°€ê²©ì€?
last_data = X.tail(1)
prediction = model.predict(last_data)
print(f"AIê°€ ì˜ˆìƒí•œ ì‚¼ì„±ì „ìì˜ ë‚´ì¼ ì¢…ê°€ëŠ”: {prediction[0]:.0f}ì›")

```
