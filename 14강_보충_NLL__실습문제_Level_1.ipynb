### ë¬¸ì œ 1.1: ê°ì„± ë¶„ì„ (Sentiment Analysis)

ê°€ì¥ ê¸°ë³¸ì´ ë˜ëŠ” ê¸°ëŠ¥ì…ë‹ˆë‹¤.



* **ì„¤ëª…:** ëª¨ë¸ ì´ë¦„ì„ ì§€ì •í•˜ì§€ ì•Šìœ¼ë©´ ê¸°ë³¸ì ìœ¼ë¡œ ê°ì„± ë¶„ì„ì— ìµœì í™”ëœ ëª¨ë¸(`distilbert-base-uncased-finetuned-sst-2-english`)ì„ ìë™ìœ¼ë¡œ ë‹¤ìš´ë¡œë“œí•˜ì—¬ ì‚¬ìš©í•©ë‹ˆë‹¤.

````python
# 'sentiment-analysis' íŒŒì´í”„ë¼ì¸ ìƒì„±
classifier = pipeline("sentiment-analysis")

result1 = classifier("I am so happy to learn stat!")
result2 = classifier("I am so heavy to learn math!")
print(f"ê²°ê³¼: {result1}")
print(f"ê²°ê³¼: {result2}")
````

### ë¬¸ì œ 1.2 - ë‹¤ë¥¸ í’€ì´ë°©ë²•
1. ë¦¬ìŠ¤íŠ¸
2. ë°ì´í„°í”„ë ˆì„ ìƒì„±
3. ëŒë‹¤ í•¨ìˆ˜ í™œìš©

````python
import pandas as pd
from transformers import pipeline

classifier = pipeline("sentiment-analysis")

# 1. ë°ì´í„°ë¥¼ ë¦¬ìŠ¤íŠ¸ë¡œ ë¬¶ê¸°
texts = [
    "I am so happy to learn stat!",
    "I am so heavy to learn math!",
    "Statistics is the grammar of science.",
    "Accounting is too much manual work."
]

# 2. ë°ì´í„°í”„ë ˆì„ ìƒì„±
df = pd.DataFrame(texts, columns=["text"])

# 3. ê°ì„± ë¶„ì„ ì ìš© (ëŒë‹¤ í•¨ìˆ˜ í™œìš©)
df['analysis'] = df['text'].apply(lambda x: classifier(x)[0])
df['label'] = df['analysis'].apply(lambda x: x['label'])
df['score'] = df['analysis'].apply(lambda x: x['score'])

# 4. ê²°ê³¼ í™•ì¸
print(df[['text', 'label', 'score']])
````
