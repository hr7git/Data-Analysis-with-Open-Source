### âœ… [ìˆ™ì œ 1 í•´ì„¤] XPathë¡œ ì›í•˜ëŠ” ë°ì´í„° ë‚šê¸°

XPathëŠ” HTMLì´ë¼ëŠ” ë³µì¡í•œ ì§€ë„ì—ì„œ íŠ¹ì • ë³´ë¬¼ì„ ì°¾ì•„ê°€ëŠ” 'ì£¼ì†Œ'ì™€ ê°™ìŠµë‹ˆë‹¤.

```python
from lxml import etree

# ì›¹ í˜ì´ì§€ ì†ŒìŠ¤(HTML)ê°€ ìˆë‹¤ê³  ê°€ì •
root = etree.fromstring(html_content)

# [í•´ì„¤]
# //a : ë¬¸ì„œ ì „ì²´ì—ì„œ <a> íƒœê·¸ë¥¼ ì°¾ì•„ë¼
# [@class='title'] : ê·¸ì¤‘ì—ì„œ class ì†ì„±ê°’ì´ 'title'ì¸ ê²ƒë§Œ ê³¨ë¼ë¼
titles = root.xpath("//a[@class='title']")

# ê²°ê³¼ ì¶”ì¶œ: .textë¡œ ê¸€ìë§Œ ê°€ì ¸ì˜¤ê³  .strip()ìœ¼ë¡œ ì•ë’¤ ê³µë°± ì œê±°
title_list = [t.text.strip() for t in titles if t.text]
print(f"ì¶”ì¶œëœ ì œëª© ê°œìˆ˜: {len(title_list)}ê°œ")
print(f"ìƒ˜í”Œ ì œëª©: {title_list[0]}")

```

### âœ… [ìˆ™ì œ 2 í•´ì„¤] ì •ê·œí‘œí˜„ì‹ìœ¼ë¡œ í…ìŠ¤íŠ¸ ì²­ì†Œí•˜ê¸°

ë°ì´í„° ë¶„ì„ì—ì„œ ê°€ì¥ ì¤‘ìš”í•œ ê²ƒì€ 'ë…¸ì´ì¦ˆ ì œê±°'ì…ë‹ˆë‹¤. íŠ¹ìˆ˜ë¬¸ìëŠ” ë¶„ì„ì˜ ì •í™•ë„ë¥¼ ë–¨ì–´ëœ¨ë¦½ë‹ˆë‹¤.

```python
import re

def clean_text(text):
    # [í•´ì„¤]
    # [^ ...] : ê´„í˜¸ ì•ˆì˜ ë‚´ìš©ì„ ì œì™¸í•œ ë‚˜ë¨¸ì§€ë¥¼ ëœ»í•¨
    # ê°€-í£ : ëª¨ë“  í•œê¸€ ë¬¸ì°¨
    # \s : ê³µë°± ë¬¸ì(ë„ì–´ì“°ê¸°)
    # ì¦‰, "í•œê¸€ê³¼ ê³µë°±ì´ ì•„ë‹Œ ëª¨ë“  ê²ƒì„ ë¹ˆì¹¸('')ìœ¼ë¡œ ë°”ê¿”ë¼"ëŠ” ëœ»ì…ë‹ˆë‹¤.
    cleaned = re.sub(r"[^ê°€-í£\s]", "", text)
    return cleaned.strip()

# í…ŒìŠ¤íŠ¸
sample_text = "ì•ˆë…•í•˜ì„¸ìš”! 2024ë…„ [ê³µì§€ì‚¬í•­] ë°ì´í„° ë¶„ì„ ë§›ì§‘ì…ë‹ˆë‹¤~^^"
print(f"ì •ì œ ì „: {sample_text}")
print(f"ì •ì œ í›„: {clean_text(sample_text)}")

```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# ë‹¨ì–´ì™€ ë¹ˆë„ìˆ˜ ì˜ˆì‹œ ë°ì´í„°
counts = {'íŒŒì´ì¬': 25, 'ë°ì´í„°': 18, 'ë¶„ì„': 15, 'íšŒê³„': 12, 'ê²½ì œ': 10}

# [í•´ì„¤]
# 1. WordCloud ê°ì²´ ìƒì„± (í•œê¸€ í°íŠ¸ ì„¤ì •ì´ ì¤‘ìš”í•©ë‹ˆë‹¤)
wc = WordCloud(
    font_path='NanumGothic.ttf', # ì½”ë© í™˜ê²½ì— ë§ëŠ” í°íŠ¸ ê²½ë¡œ
    background_color='white',
    width=800,
    height=400
)

# 2. ë¹ˆë„ìˆ˜ ë°ì´í„°ë¥¼ ë„£ì–´ êµ¬ë¦„ ìƒì„±
cloud = wc.generate_from_frequencies(counts)

# 3. í™”ë©´ ì¶œë ¥
plt.figure(figsize=(10, 5))
plt.imshow(cloud, interpolation='bilinear') # ì´ë¯¸ì§€ë¥¼ ë¶€ë“œëŸ½ê²Œ ì¶œë ¥
plt.axis('off') # ê°€ë¡œì„¸ë¡œ ì¶• ìˆ«ì ìˆ¨ê¸°ê¸°
plt.show()

```
