```python
url = 'http://openapi.seoul.go.kr:8088/sample/json/energyUseDataSummaryInfo/1/5/'
api_key ='6f5049625*********6b594e'



def get_year_months():
    year_months = []
    for year in range(2023, 2024+1):    # 2015대신에 2020년 으로 데이터 줄여서
        for month in range(1,12+1):
            year_months.append(f'{year}{month:02d}')
    return year_months

year_months_list = get_year_months()


print(type(year_months_list))
# print(year_months_list)
```

```python
import requests
import json
import pandas as pd

#  param 적용이 안되는데 ???
params = {
    'KEY': api_key,
    # 'TYPE': 'json',
    # 'SERVICE': 'energyUseDataSummaryInfo',
    # 'START_INDEX': '1',
    # 'END_INDEX': '100',
    # 'sidoName': '서울',
    # 'YEAR': '2024',  # 이것으로 test
    'MON': '01'
    }

sum_data = []

for year_month in get_year_months():
    year = year_month[:4]
    month = year_month[4:]
    new_url = f'{url}{year}/{month}/'

    try:
        response = requests.get(new_url, params=params, timeout=5)
        response.raise_for_status()

        # data = response.json()['response']['body']['items']
        data = response.json()
        sum_data.append(data)  # data 를 append

        # print('API 응답:',data)



    except requests.exceptions.RequestException as e:
        print(f"API 호출실패: {year_month}, {e}")
    except json.JSONDecodeError:
        print(f"JSON 파싱 실패 : {year_month}, {e}")



```

```python
print(sum_data)
```

```python
df = pd.DataFrame(sum_data)
df

## DataFrame 인덱스 출력
print(df.index)
## DataFrame 컬럼 출력
print(df.columns)
## DataFrame 행 출력
# print(df.values.tolist())
# print(df.values)
## DataFrame 값 출력
print(df.values.flatten())
```

```python
#  데이터 저장
df.to_csv('seoul_data.csv', index=False, encoding='utf-8')

## JSON 형식 데이터 저장
df.to_json('seoul_data.json', indent=4, force_ascii=False)

## EXCEL 형식 데이터 저장
df.to_excel('seoul_data.xlsx', index=False)
## HTML 형식 데이터 저장
df.to_html('seoul_data.html')
```
